```python
%matplotlib inline

import matplotlib
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import pandas as pd
```

```python
# Labeled data into a Pandas (awesome) DataFrame
data = pd.read_csv('data/train.csv')
```

```python
print "(Rows x Cols) :", data.shape 
# Some elements
data.head(10)
```

```python
# cross validation
features = ['Descript', 'DayOfWeek', 'PdDistrict', 'Resolution', 'Address', 'X', 'Y']
text_features = ['Descript', 'DayOfWeek', 'PdDistrict', 'Resolution']
target = ['Category']

# encode textual features to numbers
from sklearn import preprocessing
preprocessed_data = pd.DataFrame()
for f in (text_features + target):
    le = preprocessing.LabelEncoder()
    le.fit(data[f].drop_duplicates())
    preprocessed_data[f] = le.transform(data[f]) 


test_size = 0.4

from sklearn import cross_validation
X_train, X_test, y_train, y_test = cross_validation.train_test_split(preprocessed_data[text_features], 
                                                                     preprocessed_data[target], 
                                                                     test_size=test_size, random_state=1)
```
